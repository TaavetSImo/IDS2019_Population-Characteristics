```python
import pandas as pd
import numpy as np
```

```python
my_cols = [str(i) for i in range(65)] # create some row names
data = pd.read_csv('Urbbpop.csv',names=my_cols,header=None, engine="python")

df = data.copy()

df1 = df.iloc[2:]
df1.reset_index(drop=True, inplace=True)



#dropime ebavajalikud indikaatorid 
df1.drop(df1.iloc[:, 1:4], inplace = True, axis = 1)


df1.head()
```

```python
df2 = df1.T.set_index(0).T


df2.head()
```

```python
#Transpose DateFrame - Columns to rows
df_t = df2.T


#1st row to header
df_t.index.names = [None]
new_header = df_t.iloc[0]
df_t = df_t[1:]
df_t.columns = new_header


df_t.head(10)
```

```python
new = df_t.filter(['Norway','Estonia','Poland',
                   'Australia','Vanuatu','New Zealand',
                   'China','Vietnam','Georgia'], axis=1)

new.tail()
```

```python
new.to_csv('noo.csv')
```

```python
cols_plot = ['Norway','Estonia','Poland',
                   'Australia','Vanuatu','New Zealand',
                   'China','Vietnam','Georgia']
axes = new[cols_plot].plot(marker='.', alpha=0.5, figsize=(20, 18), subplots=True)
for ax in axes:
    ax.set_ylabel('Urban Population')
```
