```python
import pandas as pd
```

```python
data = pd.read_csv('co2em.csv')
df = data.copy()


df1 = df.iloc[3:]
df1.reset_index(drop=True, inplace=True)

#dropime ebavajalikud indikaatorid 
df1.drop(df1.iloc[:, 1:4], inplace = True, axis = 1)


#Valime column headeri
df2 = df1.T.set_index(0).T


df2.head()
```

```python
#Transpose DateFrame - Columns to rows
df_t = df2.T


#1st row to header
df_t.index.names = [None]
new_header = df_t.iloc[0]
df_t = df_t[1:]
df_t.columns = new_header


df_t.head(10)
```

```python
#New dataframe whith 9 selected countries
new = df_t.filter(['Australia','Belgium','Bulgaria',
                   'Australia','Belgium','Bulgaria',
                   'Australia','Belgium','Bulgaria'], axis=1)

new.head(10)
```
